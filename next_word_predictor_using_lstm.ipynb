```python
## preparing inpute sequence 
input_sequence=[]
for sentence in faqs.split('\n'):
    tokenized_sentence=tokenizer.texts_to_sequences([sentence])[0]
    for i in range(1,len(tokenized_sentence)):
        input_sequence.append(tokenized_sentence[:i+1])
    
```

```python
# doing zero padding for making length same 
max_l=max ([len(x) for x in input_sequence])
max_l
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_ip_sequence=pad_sequences(input_sequence,maxlen=max_l,padding='pre')
padded_ip_sequence
# now first 56 are input and   57th  is output 
```

```python
# converting y to one hot encode 
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=283) ## not it making number= index and at that index value is 1. if take num_classes=282 then 0 to 281 so token 282 well never one hot encode so used 283 so that 282 also one hot encode 
y.shape
## now we have prepared x ,y for training 

```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense



```

```python
model = Sequential()
model.add(Embedding(283,100,input_shape=(56,)))
# model.add(LSTM(150,return_sequences=True))
model.add(LSTM(150))
model.add(Dense(283,activation='softmax'))
```

```python
model.compile(loss="categorical_crossentropy",optimizer='adam',metrics=['accuracy'])
```

```python
def train_existing_model_on_new_text(model, tokenizer, new_text, max_seq_len=56, epochs=1):
    from tensorflow.keras.preprocessing.sequence import pad_sequences
    import numpy as np
    from tensorflow.keras.utils import to_categorical

    # 1️⃣ Tokenize new text
    token_list = tokenizer.texts_to_sequences([new_text])[0]
    
    # Ignore if too short
    if len(token_list) < 2:
        print("Text too short to train.")
        return model

    # 2️⃣ Create new training samples from the new text
    input_sequences = []
    for i in range(1, len(token_list)):
        n_gram_seq = token_list[:i+1]
        input_sequences.append(n_gram_seq)

    # 3️⃣ Pad sequences
    input_sequences = np.array(pad_sequences(input_sequences, maxlen=max_seq_len, padding='pre'))

    X = input_sequences[:, :-1]
    y = input_sequences[:, -1]

    # 4️⃣ One-hot encode output
    y = to_categorical(y, num_classes=len(tokenizer.word_index) + 1)

    # 5️⃣ Fine-tune existing model
    print("Training model on new input...")
    model.fit(X, y, epochs=epochs, verbose=1)

    print("Model fine-tuned successfully!")
    return model
```

```python
new_sentence = "This is a brand new example"
model = train_existing_model_on_new_text(model, tokenizer, new_sentence, max_seq_len=56, epochs=1)
```

```python
import tkinter as tk
from tkinter import StringVar
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

# Assuming your model and tokenizer are already loaded:
# model, tokenizer should be defined as in your code above

# --- Prediction Function ---
def predict_next_word():
    text = entry_var.get()
    if not text.strip():
        return
    
    # Tokenize input
    token_text = tokenizer.texts_to_sequences([text])[0]
    padded = pad_sequences([token_text], maxlen=56, padding='pre')
    
    # Predict
    pred = model.predict(padded, verbose=0)
    pos = np.argmax(pred)
    
    # Find word with that index
    next_word = ""
    for word, index in tokenizer.word_index.items():
        if index == pos:
            next_word = word
            break
    
    # Update suggestion label
    if next_word:
        suggestion_var.set(f"Suggested: {next_word}")
    else:
        suggestion_var.set("No suggestion found")

# --- GUI Setup ---
root = tk.Tk()
root.title("Next Word Predictor")
root.geometry("700x200")
root.config(bg="white")

entry_var = StringVar()
suggestion_var = StringVar()

# --- Search bar (like Google) ---
entry = tk.Entry(root, textvariable=entry_var, font=("Arial", 18), width=50, bd=2, relief="solid")
entry.pack(pady=30)
entry.bind("<Return>", lambda event: predict_next_word())

# --- Suggestion label below ---
suggestion_label = tk.Label(root, textvariable=suggestion_var, font=("Arial", 14), fg="gray", bg="white")
suggestion_label.pack()

# --- Predict button (optional) ---
predict_button = tk.Button(root, text="Predict Next Word", command=predict_next_word,
                           font=("Arial", 12), bg="#4285F4", fg="white", relief="flat", padx=10, pady=5)
predict_button.pack(pady=10)

root.mainloop()
```
